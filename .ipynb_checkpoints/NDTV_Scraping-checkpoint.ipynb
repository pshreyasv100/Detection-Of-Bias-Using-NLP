```python
import re
from urllib.request import urlopen
import numpy as np
from bs4 import BeautifulSoup
```

```python
blogAddress = 'https://www.ndtv.com/topic/India-US-relations'

def articleUrls(blogAddress):
    html = urlopen(blogAddress).read()
    soup = BeautifulSoup(html,"lxml")
    links = soup.findAll("a")
    #print(links)
    t = set([ link["href"] for link in links if link.has_attr('href')])
    l = []  
    for s in t:
        #a = s.split('/')
        #print(str(len(a))+str(a)+'\n')
        if ('india' in s or 'world' in s) and 'https://www.ndtv' in s:
            l.append(s)
    return l
    
    return t
l = articleUrls(blogAddress)
print(l)
```

```python
clean = lambda s: str(re.sub('[\W_]+', ' ', s))
def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

```python
def getText(articleUrl):
    try:
        html = urlopen(articleUrl).read()
        soup = BeautifulSoup(html,"lxml")
        article = soup.body.findAll("div", {"id": "ins_storybody"})
        print(articleUrl)
        article1=cleanhtml(str(article))
        article1=article1.replace('\r',' ')
        article1=article1.replace('\s',' ')
        article1=article1.replace('\n',' ')
        article1=article1.replace('\xa0',' ')
        article1=article1.replace("\'","'")
        article1=article1.strip()
        #rticle1.replace('\','')
        #print(article1)
        return article1
    except:
        print("Skipping\n")
```

```python
blogs=[]
for i in l:
    text = getText(i)
    text=str(text)
    blogs.append(text)
```

```python
for i in blogs:
    print(i+"\n\n")
```
